# For this project, we will ignore Feature Selection - because we will make use of all columns and no need to merge columns in this case.

```python
# Extracting both input and output cells
X = df.iloc[:,0:2] # Extracting input cells - need all rows, and only 0th and 1st index columns
```

# We will scale values after train, test, split

# Train, test and split

```python
from sklearn.model_selection import train_test_split
```

```python
train_test_split(X, Y, test_size = 0.1) # X, Y are input and output cells and 0.1 means 90% for training and 10% for testing)
```

# X of training set, X of test set, Y of training set and Y of test set in order above ouput

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.1)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

```python
Y_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train) # We scale values between -1 and 1.
```

```python
X_train
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

# Next Train model

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train, Y_train) #model training
```

# Next Model Evaluation

```python
clf.predict(X_test)
```

```python
Y_test
```

```python
Y_pred=clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(Y_test, Y_pred)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, Y_train.values, clf=clf, legend=2)
```
